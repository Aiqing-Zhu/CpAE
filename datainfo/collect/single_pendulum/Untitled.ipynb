```python
from scipy.integrate import solve_ivp
import numpy as np
rng = np.random.default_rng(2)

num_seq=1200
num_frm=60
    # parameters
m = 1.0
l = 0.5
g = 9.81
lam = 0.2
dt = 2.0 / 60
t_eval = np.arange(num_frm) * dt

# solve equations of motion
# y = [theta, vel_theta]
f = lambda t, y: [y[1], -3*g/(2*l) * np.sin(y[0]) - lam*y[1]]
initial_state = [rng.uniform(0, np.pi/2), rng.uniform(-5, 5)]

# initial_state=[np.pi/2, -5]
sol = solve_ivp(f, [t_eval[0], t_eval[-1]], initial_state, t_eval=t_eval, rtol=1e-6)
```

```python
import matplotlib.pyplot as plt
print(sol)
print(sol['y'].shape)
plt.plot(sol['y'][0],sol['y'][1])
```
